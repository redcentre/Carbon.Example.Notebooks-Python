### Carbon Python Sample No.01 - Simple Report

An array of values is passed as a request into the Plutonium web service. The Carbon cross-tabulation engine response contains JSON in the exact shape required for conversion into a pandas dataframe.

```python
import requests
import json
import pandas
import matplotlib.pyplot as plt
from pyparsing import line

postdata={
  "id": "10000010",
  "password": "free",
  "name": "Example 02",
  "customerName": "rcsruby",
  "jobName": "demo",
  "top": "Age",
  "side": "Region",
  "filter": None,
  "weight": None,
  "columnPercents": False,
  "formatType": 1,
  "overrideLicensingbaseAddress": None,
  "skipCache": False
}
response = requests.post("https://rcsapps.azurewebsites.net/plutonium/python/crosstab/gentab", json=postdata)
dfjson = response.json()
df=pandas.DataFrame(dfjson)
df
```
